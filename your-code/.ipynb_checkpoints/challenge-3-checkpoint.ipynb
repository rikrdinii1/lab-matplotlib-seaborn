#### Use the appropriate visulalization to show the distribution of the column `Age`

```python
# your code here
plt.hist(titanic['Age'],bins=50,color='red',alpha=.6)
plt.title('Age')
plt.show;
```

#### Use subplots to plot the distribution of the `Age` variable with bins equal to `10`, `20` and `50`.

```python
# your code here
fig,axes = plt.subplots(nrows = 1, ncols = 3)

plt.sca(axes[0])
titanic['Age'].plot(kind='hist',bins=10,color='purple',alpha=.6)
plt.title('10 Bins')
plt.xlabel('Age')

plt.sca(axes[1])
titanic['Age'].plot(kind='hist',bins=20,color='red',alpha=.6)
plt.title('20 Bins')
plt.xlabel('Age')

plt.sca(axes[2])
titanic['Age'].plot(kind='hist',bins=50,color='orange',alpha=.6)
plt.title('50 Bins')
plt.xlabel('Age')
```

#### Use the appropriate plot to visualize the column `Gender`. There are 2 ways of doing it. Do it both ways.

*Hint: Use matplotlib and seaborn*

```python
# your code here- 1st way
data = titanic.groupby('Gender').count()

plt.bar(height=data['PassengerId'],x=data.index,alpha=.6)
plt.title('Gender')
plt.show;
```

```python
# your code here- 2nd way
sns.countplot(x='Gender',data=titanic,alpha=.6);
```

#### Use the appropriate plot to visualize the column `Pclass`

#### We suspect that there is a linear relationship between `Fare` and `Age`. Use the appropriate plot to show the relationship between these 2 features. There are 2 ways, please do it both ways.

*Hint: One of the ways involves using Seaborn*

```python
#your code-1st way
x=titanic['Fare']
y=titanic['Age']

plt.scatter(x,y,alpha=0.6,color='green')
plt.xlabel('Fare')
plt.ylabel('Age')
plt.title('Age vs Fare');
```
